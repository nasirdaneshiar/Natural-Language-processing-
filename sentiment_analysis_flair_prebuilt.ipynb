1.initialize the model

```python
import flair
model  = flair.models.TextClassifier.load('en-sentiment')
```

```python
text = 'I like you!'
sentence = flair.data.Sentence(text)
model.predict(sentence)
sentence
```

```python
sentence.get_labels()[0].value
```

```python
sentence.get_labels()[0].score
```
