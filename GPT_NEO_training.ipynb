```python
from happytransformer import HappyGeneration
```

```python
happy_gen = HappyGeneration("GPT-NEO","EleutherAI/gpt-neo-125M")
```

## no settings

```python
result = happy_gen.generate_text("we must invest in ")
print(result.text)
```

```python
from happytransformer import GENSettings
```

## set no repeat ngram size

```python
greedy_settings = GENSettings(no_repeat_ngram_size=2,  max_length=50)
output_greedy = happy_gen.generate_text("we must invest in ", args=greedy_settings)
print(output_greedy.text)
```

## generic sampling settings

```python
generic_sampling_settings = GENSettings(do_sample = True ,top_k = 0, temperature = 0.7 ,no_repeat_ngram_size=3,  max_length=50)
result_generic = happy_gen.generate_text("we must invest in ",args = generic_sampling_settings) 
print(result_generic.text)
```

#training a model

```python
for i in range(10):
  happy_gen.train("ai.txt")
  result = happy_gen.eval("ai.txt")
  
  print(f'{i}:',result.loss)
```

## setting of training

```python
from happytransformer import  GENTrainArgs
# --------------------------------------#

args = GENTrainArgs(num_train_epochs=1) 
happy_gen.train("ai.txt", args=args)
```
